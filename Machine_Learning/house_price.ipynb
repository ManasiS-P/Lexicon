## Filter Data - Filter the dataset to show only houses with more than 2 bedrooms.

```python
bedrooms_2 = df[df['bedrooms'] > 2]
bedrooms_2
```

## Sort Data - Sort the dataset based on the sales price in ascending order.

```python
sorted_by_price = df.sort_values(by='sales_price', ascending=True)
```

```python
sorted_by_price
```

## Calculate Average Sales Price - Calculate the average sales price for all houses.    

```python
average_price = df['sales_price'].mean()
```

```python
average_price
```

## Add a New Column - Add a new column that shows the price per square meter.

```python
df['price_per_sqm'] = df['sales_price'] / df['sq_meter']
```

```python
df['price_per_sqm']
```

## Group Data - Group the dataset by neighborhood and calculate the average sales price for each neighborhood.

```python
avg_price_by_neighborhood = df.groupby('neighborhood')['sales_price'].mean().reset_index()
```

```python
avg_price_by_neighborhood
```

## Visualize Data - Create a simple bar plot that shows the average sales price per neighborhood.

```python
%pip install matplotlib
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
avg_price_by_neighborhood.plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('Average Sales Price per Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Average Sales Price')
plt.show()
```

## Find Maximum Values- Find the house with the highest sales price.

```python
max_value = df['sales_price'].max()
```
